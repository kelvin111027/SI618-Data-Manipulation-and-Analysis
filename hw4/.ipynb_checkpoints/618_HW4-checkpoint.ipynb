### Q4. For 10 Points: Plot the pairs of different ability points (HP, Attack, Sp. Attack, Defense, etc.).

```python
pok = pd.read_csv('Pokemon.csv')
pok.columns
```

```python
pok_ability = pok[['Type 1','HP','Attack','Defense',\
    'Sp. Atk','Sp. Def','Speed']].copy()
pok_ability.head()
```

```python
sns.heatmap(pok_ability.corr(), annot = True)
```

From the heatmap, the Sp.Atk ~ Sp.Def pair and Defense ~ Sp.Def pair have the most correlation coefficient (0.51). The Defense ~ Speed pair has the least correlation coefficient (0.015)

### Q6. For 15 Points: Explore how the Pokémon in each generation differ from each other?

* Do you think designers of *Pokémon* tried to address different distributions of ability points in each generation?

```python
pok_ability_gen = pok[['Type 1','HP','Attack','Defense','Sp. Atk',\
    'Sp. Def','Speed', 'Generation']].copy()
pok_ability_gen.head()
```

```python
sns.pairplot(pok_ability_gen, hue = 'Generation')
```

The distributions of different ability points among generations are represented by pairplot above (differentiated by colors). It seems like the fourth generation as a whole has slightly better performance on Speed and Defense than the rest generations. However, overall, I think the difference of distributions among generations are pretty small because their distributions are pretty much overlapped. So that I don't think the designers of Pokemon tried to address different distributions of ability points in each generation.
